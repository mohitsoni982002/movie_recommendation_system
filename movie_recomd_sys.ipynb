```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
from ast import literal_eval
def convert(obj):
  l = []
  for i in  ast.literal_eval(obj):
    l.append(i['name'])

  return l
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
def convert3(obj):
  l = []
  counter = 0
  for i in  ast.literal_eval(obj):
     if counter !=3:
          l.append(i['name'])
          counter +=1
  return l
```

```python
movie["cast"]=movie['cast'].apply(convert3)
```

```python
movie.head()
```

```python
def fatch_director(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      l.append(i['name'])
      break

  return l
```

```python
movie['crew']=movie['crew'].apply(fatch_director)
```

```python
movie['overview'] = movie['overview'].astype(str)
```

```python
movie['overview'] = movie['overview'].apply(lambda x: x.split())
```

Here we apply transformation for distinct identification of persons (remove spaces between names and other paragraphs or lines)

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

## We show the similarity on the basis of tags so we can recommnad the movies

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_movie['tags'][0]
```

```python
new_movie['tags'][1]
```

After that we conert this text (tags) into vectors this process called textvectorization there are many techniques for textvectorization
beg of words , tfidf , word to vek....... but here we are using beg of words technique



```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_movie['tags']).toarray()  ## spars matrix to numpy array
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

```python
stem("in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d action adventure fantasy sciencefiction samworthington zoesaldana sigourneyweaver jamescameron")
```

```python
new_movie['tags'] = new_movie['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(vectors)
```

```python
similarity_score[0]
```

when we perform sorting in similarity_score we loss its orignal index value so we use enumerat function to hold the index value of movies

```python
sorted(list(enumerate(similarity_score[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_movie[new_movie['title'] == movie].index[0]
  distances = similarity_score[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_movie.iloc[i[0]].title)
   ## print(i[0])
```

```python
recommend('Harry Potter and the Half-Blood Prince')
```
